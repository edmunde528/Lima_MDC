```python
import pandas as pd
```

```python
df1 = pd.read_csv('region_pops.csv')
df2 = pd.read_csv('pop_listings_by_region.csv')
df2.set_index(['Region'], inplace=True)
```

```python
df_grouped = pd.DataFrame(columns = ['Region','Infant', 'Nursery', 'Primary School', 'Secondary School', 'Adult'])
df_grouped['Region']=df2.index
df_grouped['Infant'] = [df2[[f'Age {i}' for i in range(0,2)]].groupby(['Region']).sum().iloc[i].sum() for i in range(df2[[f'Age {i}' for i in range(0,2)]].groupby(['Region']).sum().shape[0])]
df_grouped['Nursery'] = [df2[[f'Age {i}' for i in range(2,5)]].groupby(['Region']).sum().iloc[i].sum() for i in range(df2[[f'Age {i}' for i in range(2,5)]].groupby(['Region']).sum().shape[0])]
df_grouped['Primary School'] = [df2[[f'Age {i}' for i in range(5,12)]].groupby(['Region']).sum().iloc[i].sum() for i in range(df2[[f'Age {i}' for i in range(5,12)]].groupby(['Region']).sum().shape[0])]
df_grouped['Secondary School'] = [df2[[f'Age {i}' for i in range(12,17)]].groupby(['Region']).sum().iloc[i].sum() for i in range(df2[[f'Age {i}' for i in range(12,17)]].groupby(['Region']).sum().shape[0])]
df_grouped.set_index(['Region'], inplace=True)
df_grouped['Adult'] = [df1['population'][i] - df_grouped.iloc[i].sum() for i in range(df_grouped.shape[0])]
df_grouped
```

```python
df_sorted = pd.DataFrame(columns = ['Region','Age Group','Total'])
df_sort = pd.DataFrame(columns = ['Region','Age Group','Total'])
for i in df_grouped.columns:
    df_sort['Region'] = df_grouped.index
    df_sort['Age Group'] = i
    df_sort['Total'] =[j for j in df_grouped[i]]
    df_sorted = pd.concat([df_sorted,df_sort], axis=0)

df_sorted
```

```python
df_rates = pd.read_csv('region_rates.csv')
df_merge = pd.merge(df_sorted,df_rates,on='Region', how='inner')
df_merge['RLI Total'] = [round(df_merge['Total'][j]*df_merge['Adult BHC'][j]) if df_merge['Age Group'][j] == 'Adult' else round(df_merge['Total'][j]*df_merge['Child BHC'][j]) for j in range(df_merge.shape[0])]
df_merge['Poverty Rate'] = [df_merge['Adult BHC'][j] if df_merge['Age Group'][j] == 'Adult' else df_merge['Child BHC'][j] for j in range(df_merge.shape[0])]
```

```python
df_merge.drop(columns=['Adult BHC', 'Adult AHC', 'Child BHC', 'Child AHC'], inplace=True)
```

```python
df_merge.to_csv('pov_rate_by_region.csv', index=False)
```
